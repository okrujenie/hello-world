```
import requests
import re
import time
```

```
# ввод запроса и создание части url для поиска

question = input()

#q = re.sub(' ', '+', question)

i = 0

url = 'https://www.google.ru/search?hl=ru&q=' + question  + '&start=' + str(i*10)

headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36'}
```

```
print(f'сайт krugobaikal.ru в поиске Google по запросу "{question}":')
for i in range (5):
    next_url = url + '&start=' + str(i*10)
    search_result = requests.get(next_url, headers=headers).text
    result = re.findall('krugobaikal.ru', search_result)
    time.sleep (3)
    if result:        
        print(f'находится НА {i + 1} СТРАНИЦЕ')
        break
    else:
        print(f'на {i + 1} странице не найден ' + str(requests.get(url + '&start=' + str(i*10))))
```
